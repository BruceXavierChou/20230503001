```python
# 處理時間: 秒數 -> 格林威治時區
from datetime import datetime
str(datetime.utcfromtimestamp (1140825600))
```

```python
# 轉換流程 -> 遞給整行 -> 每一個走過流程
def timetrans(s):
    return str(datetime.utcfromtimestamp(s))
#apply: apply(流程名字) 對每一個執行一次流程
df["filmutc"] = df["film_date"].apply(timetrans)
df
```

```python
# 篩選操作: 原理 df[ [跟你的列一樣多的Boolean] ]
df.head(3)[ [False, True, True] ]
```

```python
# 篩選: 轉換操作 -> 2550 True False
def fflow(n):
    if n > 3000 and n < 5000:
        return True
    else:
        return False
f = df["comments"].apply(fflow)
df[f]
```

```python
# eval() -> 把傳入的字串當成一個python程式再執行
# eval("print(3+2)") -> 5
a = eval("[1, 2, 3]")
# a = [1, 2, 3]
a
```

```python
def fflow(s):
    taglist = eval(s)
    if "computers" in taglist:
        return True
    else:
        return False
df[df["tags"].apply(fflow)]
```
